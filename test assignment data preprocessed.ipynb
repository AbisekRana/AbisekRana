```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import tensorflow as tf 
import sklearn
```

```python
raw_data = pd.read_csv('customer_churn_large_dataset.csv')
raw_data
```

```python
data = raw_data.drop(['CustomerID'],axis=1)
```

```python
data = data.drop(['Name'],axis=1)
```

```python
data = data.drop(['Gender'],axis=1)
data.columns.values
data
```

```python
location_columns = pd.get_dummies(data['Location'])
```

```python
location_columns
```

```python
location_columns.columns.values
```

```python
l1 = location_columns.iloc[:,:1].max(axis=1)
l1
```

```python
l2 = location_columns.iloc[:,1:2].max(axis=1)
l2
```

```python
l3 = location_columns.iloc[:,2:3].max(axis=1)
l3
```

```python
l4 = location_columns.iloc[:,3:4].max(axis=1)
l4
```

```python
l5 = location_columns.iloc[:,4:5].max(axis=1)
l5
```

```python
data = pd.concat([data,l1,l2,l3,l4,l5],axis=1)
```

```python
data.columns.values
```

```python
data.columns = column_names
```
